```python
def predict_full_result(
    ripeness, weight, temperature, humidity, storage_days,
    number_of_mangoes, destination_country, transport_mode, transport_days
):
    # Create input for Model 1
    input_model1 = pd.DataFrame([{
        "batch_ripeness_level": ripeness,
        "batch_avg_weight_g": weight,
        "temperature_C": temperature,
        "humidity_%": humidity,
        "storage_days_before_shipping": storage_days
    }])

    # Predict shelf life
    shelf_life = model1.predict(input_model1)[0]

    # Check transport safety
    if shelf_life < transport_days:
        risk_level = "High Risk (Shelf Life < Transport Duration)"
    else:
        risk_level = "Safe (Shelf Life ≥ Transport Duration)"

    # Create input for Model 2
    # Note: 'box_type' and 'padding_type' are hardcoded as 'Unknown' for now,
    # but can be passed as parameters if more data is available.
    input_model2 = pd.DataFrame([{
        "number_of_mangoes": number_of_mangoes,
        "batch_ripeness_level": ripeness,
        "batch_avg_weight_g": weight,
        "destination_country": destination_country,
        "transport_mode": transport_mode,
        "transport_days": transport_days,
        "box_type": "Unknown",
        "padding_type": "Unknown"
    }])

    # Predict expected damage
    damage = model2.predict(input_model2)[0]

    # Build final recommendation
    final_output = {
        "Shelf Life (days)": round(shelf_life, 2),
        "Transport Days": transport_days,
        "Risk Level": risk_level,
        "Predicted Damage (%)": round(damage, 2)
    }

    return final_output
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score
from xgboost import XGBRegressor, XGBClassifier

# Load dataset
df2 = pd.read_excel("model2_srilanka_1500_rows.xlsx")
df2 = df2.dropna()

print("Rows:", len(df2))
print(df2.head())
```

```python
feature_cols = [
    "number_of_mangoes",
    "batch_ripeness_level",
    "batch_avg_weight_g",
    "destination_country",
    "transport_mode",
    "transport_days"
]

X = df2[feature_cols]
```

```python
y_damage = df2["expected_damage_%"]

categorical_features = ["batch_ripeness_level", "destination_country", "transport_mode"]
numeric_features = ["number_of_mangoes", "batch_avg_weight_g", "transport_days"]

preprocessor = ColumnTransformer(
    [
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),
        ("num", "passthrough", numeric_features)
    ]
)

model_damage = Pipeline([
    ("preprocess", preprocessor),
    ("xgb", XGBRegressor(
        n_estimators=500,
        learning_rate=0.1,
        max_depth=8,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    ))
])

X_train, X_test, y_train, y_test = train_test_split(X, y_damage, test_size=0.2, random_state=42)

model_damage.fit(X_train, y_train)
pred_damage = model_damage.predict(X_test)

print("Damage MAE:", mean_absolute_error(y_test, pred_damage))
print("Damage R2:", r2_score(y_test, pred_damage))
```

```python
from sklearn.preprocessing import LabelEncoder

y_box = df2["box_type"]

# Encode the target variable y_box
label_encoder = LabelEncoder()
y_box_encoded = label_encoder.fit_transform(y_box)

model_box = Pipeline([
    ("preprocess", preprocessor),
    ("xgb", XGBClassifier(
        n_estimators=400,
        learning_rate=0.1,
        max_depth=8,
        subsample=0.9,
        colsample_bytree=0.7,
        random_state=42
    ))
])

X_train2, X_test2, y_train2_encoded, y_test2_encoded = train_test_split(X, y_box_encoded, test_size=0.2, random_state=42)

model_box.fit(X_train2, y_train2_encoded)
pred_box_encoded = model_box.predict(X_test2)

# Decode predictions for accuracy calculation (optional, can also use encoded labels)
# pred_box = label_encoder.inverse_transform(pred_box_encoded)

print("Box Type Accuracy:", accuracy_score(y_test2_encoded, pred_box_encoded))
```

```python
from sklearn.preprocessing import LabelEncoder

y_padding = df2["padding_type"]

# Encode the target variable y_padding
label_encoder_padding = LabelEncoder()
y_padding_encoded = label_encoder_padding.fit_transform(y_padding)

model_padding = Pipeline([
    ("preprocess", preprocessor),
    ("xgb", XGBClassifier(
        n_estimators=400,
        learning_rate=0.1,
        max_depth=8,
        subsample=0.9,
        colsample_bytree=0.7,
        random_state=42
    ))
])

X_train3, X_test3, y_train3_encoded, y_test3_encoded = train_test_split(X, y_padding_encoded, test_size=0.2, random_state=42)

model_padding.fit(X_train3, y_train3_encoded)
pred_padding_encoded = model_padding.predict(X_test3)

# Decode predictions for accuracy calculation (optional, can also use encoded labels)
# pred_padding = label_encoder_padding.inverse_transform(pred_padding_encoded)

print("Padding Type Accuracy:", accuracy_score(y_test3_encoded, pred_padding_encoded))
```

```python
def full_packaging_recommendation(
    shelf_life,
    number_of_mangoes, ripeness, weight,
    destination_country, transport_mode, transport_days,
    temperature, humidity, storage_days
):

    # Model 2 input
    input2 = pd.DataFrame([{
        "number_of_mangoes": number_of_mangoes,
        "batch_ripeness_level": ripeness,
        "batch_avg_weight_g": weight,
        "destination_country": destination_country,
        "transport_mode": transport_mode,
        "transport_days": transport_days
    }])

    # Predictions
    predicted_damage = model_damage.predict(input2)[0]
    predicted_box = model_box.predict(input2)[0]
    predicted_padding = model_padding.predict(input2)[0]

    # Legal Check
    country_rules = legal_rules.get(destination_country, None)
    legal_status = "Unknown"
    legality_comment = ""

    if country_rules:
        if predicted_box in country_rules["banned"]:
            legal_status = "Illegal"
            predicted_box = country_rules["allowed"][0]
            legality_comment = "Box type changed to comply with destination regulations."
        else:
            legal_status = "Legal"
    else:
        legal_status = "Unknown"
        legality_comment = "No legal data found for this country."

    # Eco Check
    eco_status = eco_rules.get(predicted_box, "Unknown")

    # Risk Check
    if shelf_life < transport_days:
        risk_msg = "⚠ HIGH RISK: Shelf life is lower than transport days!"
    else:
        risk_msg = "✓ Safe: Shelf life is acceptable for transport."

    # Build Recommendation
    return {
        "Shelf Life (days)": round(shelf_life, 2),
        "Transport Days": transport_days,
        "Risk": risk_msg,
        "Predicted Damage (%)": round(predicted_damage, 2),
        "Recommended Box Type": predicted_box,
        "Recommended Padding": predicted_padding,
        "Eco Status": eco_status,
        "Legal Status": legal_status,
        "Legal Note": legality_comment,
        "Final Recommendation": f"Use {predicted_box} with {predicted_padding}. Expected damage {round(predicted_damage,2)}%. Eco: {eco_status}. Legal: {legal_status}."
    }
```

```python
def model2_full_recommendation(
    shelf_life,
    number_of_mangoes, ripeness, weight,
    destination_country, transport_mode, transport_days
):

    # MODEL INPUT
    input_df = pd.DataFrame([{
        "number_of_mangoes": number_of_mangoes,
        "batch_ripeness_level": ripeness,
        "batch_avg_weight_g": weight,
        "destination_country": destination_country,
        "transport_mode": transport_mode,
        "transport_days": transport_days
    }])

    # PREDICTIONS
    damage = float(model_damage.predict(input_df)[0])

    # decode labels
    box_idx = model_box.predict(input_df)[0]
    padding_idx = model_padding.predict(input_df)[0]

    box_type = box_classes[box_idx]
    padding_type = padding_classes[padding_idx]

    # LEGAL CHECK
    rules = legal_rules.get(destination_country)
    legal_status = "Unknown"
    note = ""

    if rules:
        if box_type in rules["banned"]:
            old_box = box_type
            box_type = rules["allowed"][0]
            legal_status = "Illegal (changed)"
            note = f"Box type '{old_box}' was illegal. Changed to '{box_type}'."
        else:
            legal_status = "Legal"

    # ECO CHECK
    eco_status = eco_rules.get(box_type, "Unknown")

    # RISK CHECK
    risk = "⚠ HIGH RISK (Shelf life < transport duration)" if shelf_life < transport_days else "✓ Safe"

    # FINAL OUTPUT
    return {
        "Shelf Life": shelf_life,
        "Transport Days": transport_days,
        "Risk": risk,
        "Predicted Damage (%)": round(damage, 2),
        "Recommended Box": box_type,
        "Recommended Padding": padding_type,
        "Eco Status": eco_status,
        "Legal Status": legal_status,
        "Legal Note": note,
        "Final Recommendation": (
            f"Use {box_type} with {padding_type}. "
            f"Expected damage: {round(damage, 2)}%. "
            f"Eco: {eco_status}. Legal: {legal_status}."
        )
    }
```
